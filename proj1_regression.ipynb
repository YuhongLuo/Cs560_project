```python
#names = ['Jan', 'Feb']
data = pd.read_csv('/Users/yuhongluo/Desktop/project11.csv')
data.plot(kind='box', subplots=True, layout=(3,5), sharex=False, sharey=False)
plt.show()
```

```python
seaborn.heatmap(data.corr(), annot=True, square=True, vmin=-0.5, vmax=0.5)
```

```python
scatter_matrix(data)
plt.show()
```

```python
plt.figure(figsize=(16, 8))
plt.scatter(
    data['MarRainN'],
    data['FebRainN'],
    c='black'
)
plt.xlabel("March Rainfall of NYC")
plt.ylabel("Febrary Rainfall of NYC")
plt.show()
```

```python
from sklearn.linear_model import LinearRegression
X = data['FebRainN'].values.reshape(-1,1)
y = data['MarRainN'].values.reshape(-1,1)
reg = LinearRegression()
reg.fit(X, y)
print("The linear model is: Y = {:.5} + {:.5}X".format(reg.intercept_[0], reg.coef_[0][0]))
```

```python
predictions = reg.predict(X)
plt.figure(figsize=(16, 8))
plt.scatter(
    data['FebRainN'],
    data['MarRainN'],
    c='black'
)
plt.plot(
    data['FebRainN'],
    predictions,
    c='blue',
    linewidth=2
)
plt.xlabel("February Rainfall of NYC")
plt.ylabel("March Rainfall of NYC")
plt.show()
```

```python
import statsmodels.api as sm
X = data['FebRainN']
y = data['MarRainN']
X2 = sm.add_constant(X)
est = sm.OLS(y, X2)
est2 = est.fit()
print(est2.summary())
```
