```python
plt.figure(figsize=(16, 8))
plt.scatter(
    d2['AugTemSJ'],
    d2['SepTemSJ'],
    c='black'
)
plt.xlabel("Auguest Tempratrue of San Jose")
plt.ylabel("September Tempratrue of San Jose")
plt.show()
```

```python
from sklearn.linear_model import LinearRegression
X = d2['AugTemSJ'].values.reshape(-1,1)
y = d2['SepTemSJ'].values.reshape(-1,1)
reg = LinearRegression()
reg.fit(X, y)
print("The linear model is: Y = {:.5} + {:.5}X".format(reg.intercept_[0], reg.coef_[0][0]))
```

```python
predictions = reg.predict(X)
plt.figure(figsize=(16, 8))
plt.scatter(
    d2['AugTemSJ'],
    d2['SepTemSJ'],
    c='black'
)
plt.plot(
    d2['AugTemSJ'],
    predictions,
    c='blue',
    linewidth=2
)
plt.xlabel("Auguest Tempratrue of San Jose")
plt.ylabel("September Tempratrue of San Jose")
plt.show()
```

```python
import statsmodels.api as sm
X = d2['AugTemSJ']
y = d2['SepTemSJ']
X2 = sm.add_constant(X)
est = sm.OLS(y, X2)
est2 = est.fit()
print(est2.summary())
```
